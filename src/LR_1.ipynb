***Помилки даних***

***Усі унікальні значення***

```python

unique_values = {col: ds[col].unique() for col in ds.columns}
print("\nUnique values ​​in columns:")
for col, values in unique_values.items():
    print(f"{col}: {values}")
```

***Перевірка на можливі помилки даних***

```python
def check_invalid_values(df):
    invalid_values_summary = {}

    for column in df.columns:
        null_count = df[column].isnull().sum()
        question_mark_count = (df[column] == '?').sum()
        value_999_count = (df[column] == 999).sum()
        value_unknown_count = (df[column] == 'unknown').sum()
        value_dot_count = (df[column] == '.').sum()

        total_invalid = null_count + question_mark_count + value_999_count+value_dot_count+value_unknown_count

        if total_invalid > 0:
            invalid_values_summary[column] = {
                'null_count': null_count,
                'question_mark_count': question_mark_count,
                'value_999_count': value_999_count,
                'value_dot_count':value_dot_count,
                'value_unknown_count':value_unknown_count,
                'total_invalid': total_invalid
            }

    return invalid_values_summary


invalid_values_report = check_invalid_values(ds)

for column, counts in invalid_values_report.items():
    print(f"\n{column}:")
    print(f"  Null values: {counts['null_count']}")
    print(f"  '?' values: {counts['question_mark_count']}")
    print(f"  999 values: {counts['value_999_count']}")
    print(f"  dot values: {counts['value_dot_count']}")
    print(f"  unknown values: {counts['value_unknown_count']}")
    print(f"  Total invalid values: {counts['total_invalid']}")
```
